```python
import numpy as np 
import pandas as pd
def fundamentals(ticker):
    ticker = ticker + '.NS'
    return_columns = ['Trailing P/E' , 'Price/sales (ttm)' , 'Price/book (mrq)' , 'Return on equity (ttm)' ,
                   'Return on assets (ttm)' , 'Operating margin (ttm)' , 'Profit margin' ,'Total debt/equity (mrq)' , 
                   'Current ratio (mrq)']
    url = f'https://in.finance.yahoo.com/quote/{ticker}/key-statistics?p={ticker}'
    list_of_df = pd.read_html(url)

    df = pd.concat(list_of_df[0:] , axis=0)
    df.index = df[0]
    df.drop(0 ,axis=1 , inplace=True)

    r = dict()
    for col in return_columns:
        v = str(df.loc[col][1]).split('%')[0]
        val = {col :float(v)}
        r.update(val)
    r.update({'Ticker':ticker})
    row = pd.DataFrame([r.values()] , columns=r.keys())
    return row
```

```python
ticker = 'SBIN'
df = fundamentals(ticker)
```
